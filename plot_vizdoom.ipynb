```python
import matplotlib.pyplot as plt
import pandas as pd

sync_df = pd.read_csv('./sync_vizdoom_2.log', header = None, sep=" ")
async_df = pd.read_csv('./async_vizdoom.log', header = None, sep= " ")
```

```python
plt.figure(figsize=(10,6))
plt.plot(async_df[0],async_df[2], 'o--', label='Async', color = 'orange')
plt.plot(sync_df[0],sync_df[2], 's-', label ='Sync', color ='blue')

plt.xlabel('Number of Envs')
plt.ylabel('Steps per second')
plt.title('Steps per second vs number of envs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sentence = 'Life is short, eat dessert first'

dc = {s:i for i,s 
      in enumerate(sorted(sentence.replace(',', '').split()))}

print(dc)
```

```python
sorted(sentence.replace(',','').split())
```

```python
import torch

sentence_int = torch.tensor(
    [dc[s] for s in sentence.replace(',', '').split()]
)
print(sentence_int)
```

```python
vocab_size = 50_000
embed = torch.nn.Embedding(vocab_size, 3)
```

```python
embedded_sentence = embed(sentence_int).detach()
```
